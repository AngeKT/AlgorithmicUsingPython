## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python

import matplotlib.pyplot as plt
import pandas as pd

sales_data=pd.read_csv("../data/company_sales_data.csv")
profit=sales_data['total_profit']
month=sales_data['month_number']
plt.plot(month, profit)
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.title('Company profit per month')
plt.show()
```

**Read all product sales data and show it  using a multiline plot**

```python
# Let's load the data first
cream=sales_data['facecream']
wash=sales_data['facewash']
paste=sales_data['toothpaste']
soap=sales_data['bathingsoap']
poo=sales_data['shampoo']
rizer=sales_data['moisturizer']

# Let's plot now
plt.figure(figsize=(8, 3))
plt.plot(month, cream, label='Face cream Sales Data')
plt.plot(month, wash, label='Face wash Sales Data')
plt.plot(month, paste, label='Toothpaste Sales Data')
plt.plot(month, soap, label='Bathingsoap Sales Data')
plt.plot(month, poo, label='Shampoo Sales Data')
plt.plot(month, rizer, label='Moisturizer Sales Data')

# Let's label the x- and y-axes
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

# Let's add the titles and legends
plt.title('Sales data')
plt.legend(loc="upper left")
```

## Exercise 4
**Read toothpaste sales data of each month and show it using a scatter plot**
Also, add a grid in the plot. gridline style should .

**Expected output**

<img src="../data/Tooth_paste_Sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# Let's load the data we will use
month=sales_data['month_number']
paste=sales_data['toothpaste']

# Let's plot the scattered figure
fig=plt.figure(figsize=(14, 3))
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(month, paste, label='Toothpaste Sales data')
plt.grid(True, linewidth=0.5, linestyle='dashdot')

# Let's label, entitle and legend the plot
ax.set_xlabel("Month Number")
ax.set_ylabel("Number of units Sold")
ax.set_title("Toothpaste sales data")
plt.legend(loc="upper left")
plt.show()
```
